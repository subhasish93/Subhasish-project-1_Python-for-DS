```python
#Retrieve the names of top20 players based on the Overall rating.
top_20_players = data.nlargest(20, 'Overall')['Name']
print("Top 20 Players based on Overall Rating:")
print(top_20_players)
```

```python
#Generate a dataframe which should include all the information of the
#Top 20 players based on the Overall rating.
Top_20_players_df = data.nlargest(20, 'Overall')
Top_20_players_df.head(2)
```

```python
# What is the average "Age" and "Wage" of these top 20 players?
Top_20_players_df['Wage'] = pd.to_numeric(Top_20_players_df['Wage'].str.replace('[^\d]', ''), errors='coerce')

# Convert 'Age' column to numeric
Top_20_players_df['Age'] = pd.to_numeric(Top_20_players_df['Age'], errors='coerce')

# Calculate average age and wage
average_age = Top_20_players_df['Age'].mean()
average_wage = Top_20_players_df['Wage'].mean()

# Print or use the values as needed
print("Average Age:", average_age)
print("Average Wage:", average_wage)
```

```python
#Among the top 20 players based on the Overall rating, which player has
#the highest wage? Display the name of the player with his wage.
player_highest_wage = Top_20_players_df[Top_20_players_df['Wage'] == Top_20_players_df['Wage'].max()]
print("Player with the highest wage among Top 20 Players:")
print(player_highest_wage[['Name', 'Wage']])
```

```python
#Generate a dataframe which should include the "Player name", "Club
# Name", "Wage", and 'Overall rating'.
player_club_wage_overall = data[['Name', 'Club', 'Wage', 'Overall']]
player_club_wage_overall.head()
```

```python
#find the average Overall rating for each club
Average_overall_rating = player_club_wage_overall.groupby('Club')['Overall'].mean()
Average_overall_rating
```

```python
#Display the average overall rating of Top10 Clubs using a plot
top_10_clubs = Average_overall_rating.nlargest(10)
plt.figure(figsize=(10, 6))
top_10_clubs.plot(kind='bar', color='skyblue')
plt.title('Average Overall Rating of Top 10 Clubs')
plt.xlabel('Clubs')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#What is the relationship between age and individual potential of the player?
sns.regplot(x='Age', y='Potential', data=Top_20_players_df)
plt.title('Relationship between Age and Potential')
plt.xlabel('Age')
plt.ylabel('Potential')
plt.show()
```

From this graph we conclude that with age getting older potential start decreasing.

```python
#Which features directly contribute to the wages of the players?
# Support your answer with a plot and a metric. (2 points) (hint: use
# potential, Overall, value, international reputation, and Release Clause)
selected_columns = ['Potential', 'Overall', 'Value', 'International Reputation', 'Release Clause', 'Wage']

# Create a subset DataFrame with selected columns
subset_df = Top_20_players_df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = subset_df.corr()

# Plot a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Player Features and Wages')
plt.show()
```

**Higher the correlation value higher is the performance and wages**

```python
#Find the position in the pitch where the maximum number of players
#play and the position where the minimum number of players play? Display
#it using a plot.
position_counts = Top_20_players_df['Position'].value_counts()


max_position = position_counts.idxmax()
min_position = position_counts.idxmin()


plt.figure(figsize=(12, 6))
sns.countplot(x='Position', data=Top_20_players_df, order=position_counts.index)
plt.title('Number of Players in Each Position')
plt.xlabel('Position')
plt.ylabel('Number of Players')


plt.annotate(f'Max: {max_position}',
             xy=(position_counts.idxmax(), position_counts.max()),
             xytext=(position_counts.idxmax(), position_counts.max() + 1),
             ha='center', fontsize=10, color='red',
             arrowprops=dict(facecolor='red', shrink=0.05))


plt.annotate(f'Min: {min_position}',
             xy=(position_counts.idxmin(), position_counts.min()),
             xytext=(position_counts.idxmin(), position_counts.min() + 1),
             ha='center', fontsize=10, color='blue',
             arrowprops=dict(facecolor='blue', shrink=0.05))

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#How many players are from the club 'Juventus' and the wage is greater than 200K?
juventus_players_high_wage = Top_20_players_df[(Top_20_players_df['Club'] == 'Juventus') & (Top_20_players_df['Wage'] > 200000)]

# Display information of such players
print("Players from Juventus with wage > 200K:")
print(juventus_players_high_wage)
```

```python
#Generate a data frame each unique position containing top 5 players by Overall rating for each unique position.
sorted_data = Top_20_players_df.sort_values(by='Overall', ascending=False)

# Group by 'Position' and select the top 5 players for each group
top_5_players_by_position = sorted_data.groupby('Position').apply(lambda group: group.head(5))

# Display the resulting DataFrame
print(top_5_players_by_position)
```

```python
#What is the average wage one can expect to pay for the top 5 players
#in every position? (use the data frame created in Q19)
average_wage_by_position = top_5_players_by_position.groupby('Position')['Wage'].mean()

print("Average Wage for the Top 5 Players in Every Position:")
print(average_wage_by_position)
```
